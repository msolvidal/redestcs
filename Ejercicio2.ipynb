```python
def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data
```

```python
def Diam (H):    
    Gcc = sorted(nx.connected_component_subgraphs(H), key=len, reverse=True)
    G0 = Gcc[0]
    return G0.number_of_nodes()
    
def Sacando_nodos(Grafo,GN):
    ComponenteGigante=[]
    N=0
    for g,n in GN:
        if N<61:

            N+=1
            Grafo.remove_node(n)
            ComponenteGigante.append([Diam(Grafo),N])

    return(ComponenteGigante)
```

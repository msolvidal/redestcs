```python
def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

red_delfines=ldata("data/new_dolphins.gml")
```

```python
G=nx.read_gml("data/new_dolphins.gml")
nx.draw(G, with_labels=False, font_weight='bold',node_size=10)
plt.title("Red delfines")
plt.show()
```
